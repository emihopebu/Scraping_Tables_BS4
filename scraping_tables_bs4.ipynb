```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import sqlalchemy
```

### HTTP Request

#### store website in variable

```python
url="https://www.skyscrapercenter.com/buildings"
```

```python
response=requests.get(url)
```

#### status code

```python
response
```

### Soup Object

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

### Results

```python
# starting point
results=soup.find('table', {'id':'buildingsTable'}).find('tbody').find_all('tr')
```

```python
len(results)
```

### Find your Data

```python
# name
results[0].find_all('td')[1].get_text().strip()
```

```python
# city
results[0].find_all('td')[2].get_text().strip()
```

```python
# completion_year
results[0].find_all('td')[4].get_text().strip()
```

```python
# height
results[0].find_all('td')[5].find_all('p')[0].get_text()
```

```python
# floors
results[0].find_all('td')[6].get_text().strip()
```

```python
# material
results[0].find_all('td')[7].get_text().strip()
```

```python
# function
results[0].find_all('td')[8].get_text().strip()
```

#### put everything together - For Loop

```python
name=[]
city=[]
completion_year=[]
height=[]
floors=[]
material=[]
function=[]

for result in results:
    
     #name
      name.append(result.find_all('td')[1].get_text().strip())
    
    #city
      city.append(result.find_all('td')[2].get_text().strip())
    #completion_year
      completion_year.append(result.find_all('td')[4].get_text().strip())
    #height
      height.append(result.find_all('td')[5].find_all('p')[0].get_text())
    #floors
      floors.append(result.find_all('td')[6].get_text().strip())
    #material
      material.append(result.find_all('td')[7].get_text().strip())
    #function
      function.append(result.find_all('td')[8].get_text().strip())
    
```

```python
len(floors)
```

#### Pandas Dataframe

```python
df_buildings=pd.DataFrame({"name":name,"city":city,"completion_year":completion_year,"height":height,"floors":floors,"material":material,"function":function})
```

```python
df_buildings
```

#### Export

```python
# csv
df_buildings.to_csv('buildings.csv', sep=',',encoding='utf-8-sig',index=False)
```

```python
# excel
df_buildings.to_excel('buildings-excel.xlsx',index=False)
```

```python
# json
df_buildings.to_json('buildings.json',orient='records',lines=True, force_ascii=False)
```

```python
# postgresql
engine = sqlalchemy.create_engine('postgresql://postgres:12345@localhost:5432')
df_buildings.to_sql('buildings', engine, index=False)
```
